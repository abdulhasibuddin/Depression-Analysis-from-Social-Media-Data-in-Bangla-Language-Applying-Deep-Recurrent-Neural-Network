```python
# lstm plotting validation accuracies against iterations:
smoothing_factor = 200
linestyle='-.'
linewidth = 2.5

#plt.figure(figsize=(13,8))
plt.figure(figsize=(25,15))
plt.title('Validation Accuracies for LSTM Models')
plt.xlabel('iterations')
plt.ylabel('validation accuracy')

for i in range(0,len(x_list)):
    label = "model-"+str(i+1)+" ("+str(required_iteration_list[i])+" iterations)"
    x = x_list[i]
    y = model_list[i]
    x_sm = np.array(x)
    y_sm = np.array(y)    
    x_smooth = np.linspace(x_sm.min(), x_sm.max(), smoothing_factor)
    y_smooth = spline(x, y, x_smooth)
    plt.plot(x_smooth, y_smooth, linestyle=linestyle, linewidth=linewidth, label=label)
    
plt.grid(True)
plt.legend()
plt.savefig('images\lstm_image_1_plotting _validation _accuracies _against _iterations.png', bbox_inches='tight')
plt.show()
```
