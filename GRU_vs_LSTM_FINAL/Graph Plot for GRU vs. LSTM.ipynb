```python
gru_models_test_acc_list = [59.1,70.0,67.3,74.5,69.1,52.0,61.0,75.7,70.3,57.0,61.0,74.8,69.6,56.5]
lstm_models_test_acc_list = [50.0,73.6,72.7,69.1,70.9,36.0,59.0,73.9,77.1,57.0,40.0,73.9,42.6,62.6]

print(len(gru_models_test_acc_list))
print(len(lstm_models_test_acc_list))
```

```python
x_axis_8 = []
iterations_model_8 = 560
for iter_no in range(1,iterations_model_8+1):
    if iter_no%25 == 0:
        x_axis_8.append(iter_no)
        
x_axis_9 = []
iterations_model_9 = 1880
for iter_no in range(1,iterations_model_9+1):
    if iter_no%25 == 0:
        x_axis_9.append(iter_no)
```

```python
# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = gru_models_test_acc_list
bars2 = lstm_models_test_acc_list

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(15,8))
# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='gru-models')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='lstm-models')

objects = []
for i in range(0,len(gru_models_test_acc_list)):
    object_name = "Model-"+str(i+1)
    objects.append(object_name)

# Add xticks on the middle of the group bars
plt.xlabel('Model No.', fontweight='bold')
plt.ylabel('Test Accuracy (%)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], objects)
 
# Create legend & Show graphic
plt.legend()
plt.savefig('images\compare_image_1_gru_vs_lstm_test_acc.png', bbox_inches='tight')
plt.show()
```

```python
smoothing_factor = 200
linestyle='-.'
marker = ''
markersize = 3

#plt.figure(figsize=(30,15))
plt.title('GRU vs LSTM Best Model Validation Accuracies')
plt.xlabel('iterations')
plt.ylabel('validation accuracy')

x = x_axis_8
y = gru_model_8_val_acc
plt.plot(x, y, marker=marker, markersize=markersize, linestyle=linestyle, label='gru-model-8(560 iters)')

y = lstm_model_8_val_acc
plt.plot(x, y, marker=marker, markersize=markersize, linestyle=linestyle, label='lstm-model-8(560 iters)')

x = x_axis_9
y = lstm_model_9_val_acc
plt.plot(x, y, marker=marker, markersize=markersize, linestyle=linestyle, label='lstm-model-9(1880 iters)')

plt.grid(True)
plt.legend()
#plt.savefig('images\compare_image_2_gru_vs_lstm_best_models_val_acc.png', bbox_inches='tight')
plt.show()
```

```python
smoothing_factor = 200
linestyle='-.'
marker = ''
markersize = 3

#plt.figure(figsize=(30,15))
plt.title('GRU vs LSTM Best Model Validation Accuracies')
plt.xlabel('iterations')
plt.ylabel('validation accuracy')

x = x_axis_8
y = gru_model_8_val_acc
x_sm = np.array(x)
y_sm = np.array(y)    
x_smooth = np.linspace(x_sm.min(), x_sm.max(), smoothing_factor)
y_smooth = spline(x, y, x_smooth)
plt.plot(x_smooth, y_smooth, marker=marker, markersize=markersize, linestyle=linestyle, label='gru-model-8(560 iters)')

y = lstm_model_8_val_acc
x_sm = np.array(x)
y_sm = np.array(y)    
x_smooth = np.linspace(x_sm.min(), x_sm.max(), smoothing_factor)
y_smooth = spline(x, y, x_smooth)
plt.plot(x_smooth, y_smooth, marker=marker, markersize=markersize, linestyle=linestyle, label='lstm-model-8(560 iters)')

x = x_axis_9
y = lstm_model_9_val_acc
x_sm = np.array(x)
y_sm = np.array(y)    
x_smooth = np.linspace(x_sm.min(), x_sm.max(), smoothing_factor)
y_smooth = spline(x, y, x_smooth)
plt.plot(x_smooth, y_smooth, marker=marker, markersize=markersize, linestyle=linestyle, label='lstm-model-9(1880 iters)')

plt.grid(True)
plt.legend()
plt.savefig('images\compare_image_2_gru_vs_lstm_best_models_val_acc.png', bbox_inches='tight')
plt.show()
```
